```python
dataset["Education"]=dataset['Education'].astype('category')
dataset.dtypes
dataset["JobInvolvement"]=dataset['JobInvolvement'].astype('category')
dataset.dtypes
dataset["PerformanceRating"]=dataset['PerformanceRating'].astype('category')
dataset.dtypes
dataset["Attrition"]=dataset['Attrition'].astype('category')
dataset.dtypes
```

```python
dataset["Education"]=dataset["Education"].cat.codes
dataset
dataset["JobInvolvement"]=dataset["JobInvolvement"].cat.codes
dataset
dataset["PerformanceRating"]=dataset["PerformanceRating"].cat.codes
dataset
dataset["Attrition"]=dataset["Attrition"].cat.codes
dataset
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(categorical_features = [0, max(1470)])
dataset = onehotencoder.fit_transform(dataset).toarray()
```

```python
print(dataset)
```

```python
from sklearn.model_selection import train_test_split

training_data, testing_data = train_test_split(dataset, test_size=0.2, random_state=25)
```
